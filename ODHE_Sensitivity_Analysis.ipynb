
# ODHE Ethylene Production - Techno-Economic Sensitivity Analysis

## ‚ú® Objective
Perform a sensitivity analysis to determine how variations in key economic parameters affect the Minimum Selling Price (MSP) of ethylene produced via the ODHE process.

## üì¶ Assumptions
- Plant operational time: 8000 hours/year
- Ethylene production: 91.6 tons/hour
- CAPEX: 220 million USD
- OPEX (base): 578 million USD/year (excluding ethane cost variation)

## üîë Key Variables
- Ethane Price ($/ton): 800 to 1300
- Electricity Price ($/kWh): 0.05 to 0.10
- Steam Price ($/ton): 10 to 25
- Refrigeration Price ($/kWh): 0.03 to 0.08

## üèóÔ∏è Model Formulation
The MSP is calculated as:

\$\$
\text{MSP} = \frac{\text{Total Annual Cost}}{\text{Annual Ethylene Production (tons)}}
\$\$

## üî¢ MSP vs Ethane Price


```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Data import
data = pd.read_csv('MSP_vs_Ethane.csv')

# Plot MSP vs Ethane Price
plt.figure(figsize=(8,6))
plt.plot(data['Ethane Price ($/ton)'], data['MSP ($/ton)'], marker='o')
plt.title('MSP Sensitivity to Ethane Price')
plt.xlabel('Ethane Price ($/ton)')
plt.ylabel('MSP ($/ton ethylene)')
plt.grid(True)
plt.show()

data
```


## üå™Ô∏è Tornado Chart Analysis

This plot shows which cost variables have the highest impact on MSP.


```python

# Tornado chart
deviations_sorted = {
    'Ethane Price': 616.81,
    'Electricity Price': 200.0,
    'Refrigeration Price': 150.0,
    'Steam Price': 120.0
}

plt.figure(figsize=(8,6))
plt.barh(list(deviations_sorted.keys()), list(deviations_sorted.values()), color='orange')
plt.title('Tornado Chart: MSP Sensitivity')
plt.xlabel('MSP Deviation ($/ton)')
plt.grid(axis='x')
plt.show()
```


## üéØ Conclusion
- **Ethane price** is the most dominant factor influencing MSP.
- **Electricity and Refrigeration costs** are secondary but significant.
- This model highlights key levers for cost control in the ODHE process.
